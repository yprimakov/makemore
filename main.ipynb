```python
names = open('names.txt', 'r').read().splitlines()

names[:10]
```

```python
minlen = min(len(n) for n in names)
maxlen = max(len(n) for n in names)
print(f"Shortest name length is {minlen}")
print(f"Longest name length is {maxlen}")
```

```python
# Bigram model
for n in names[:3]:
    chs = ['.'] + list(n) + ['.']  # add start and end markers
    for ch1, ch2 in zip(chs, chs[1:]):
        print(ch1, ch2)
```

```python
# Bigram model
b = {}
for n in names:
    chs = ['.'] + list(n) + ['.']           # add start and end markers
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)                 # bigram is a tuple of two characters
        b[bigram] = b.get(bigram, 0) + 1    # increment count for bigram, or set to 1 if not seen before

sorted(b.items(), key=lambda kv: -kv[1])    # created a sorted list by count, descending
```
